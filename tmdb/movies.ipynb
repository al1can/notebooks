```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
collections = movies_metadata['belongs_to_collection'].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
print(collections)
print(collections.values[0]['id'])
```

```python
for col in collections:
    if type(col) == dict:
        print(col['id'])
```

```python
movies_metadata['belongs_to_collection'] = collections.apply(lambda x: str(x['id']) if type(x) == dict else None)
```

```python
movies_metadata.drop(columns=['genres', 'production_companies', 'production_countries', 'spoken_languages'], inplace=True)
```

```python
# , 'created_at', 'updated_at'
movies_metadata = movies_metadata[['id', 'imdb_id', 'title', 'original_title', 'overview', 'poster_path', 'release_date', 'tagline', 'homepage', 'original_language', 'belongs_to_collection', 'revenue', 'runtime', 'status', 'popularity', 'adult', 'budget', 'video', 'vote_average', 'vote_count']]
```

```python
movies_metadata['created_at'] = pd.Timestamp.now()
movies_metadata['updated_at'] = pd.Timestamp.now()
```

```python
movies_metadata = movies_metadata.drop_duplicates()
```

```python
for i, overview in enumerate(movies_metadata['overview']):
    if not pd.isna(overview) and '\n' in overview:
        #print(movies_metadata.loc[i, 'overview'])
        movies_metadata.loc[i, 'overview'] = movies_metadata.loc[i, 'overview'].replace('\n', '')
```

```python
movies_metadata = movies_metadata.replace('\n', '', regex=True)
```

```python
movies_metadata.to_csv('./clean_db/movies.csv', index=False)
```
