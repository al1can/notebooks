```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
movies_metadata.loc[0]['production_companies']
```

```python
df_dict = {'id': [], 'name': []}
```

```python
columns = movies_metadata['production_companies'].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
columns
```

```python
for x, col in enumerate(columns.values):
    if not type(col) == list:
        print(col)
        columns[x] = np.NaN
```

```python
companies = []
for col in columns.values:
    if type(col) == list:
        for company in col:
            if company not in companies:
                companies.append(company)
companies
```

```python
for company in companies:
    df_dict['id'].append(company['id'])
    df_dict['name'].append(company['name'])
```

```python
df_dict
```

```python
movies_companies = pd.DataFrame.from_dict(df_dict)
movies_companies
```

```python
movies_companies.to_csv('./clean_db/movies_companies.csv', index=False)
```
