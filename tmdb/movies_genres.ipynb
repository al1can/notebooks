```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
df_dict = {'id': [], 'name': []}
```

```python
columns = movies_metadata['genres'].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
columns
```

```python
genres = []
for col in columns.values:
        for genre in col:
            if genre not in genres:
                genres.append(genre)
genres
```

```python
for genre in genres:
    df_dict['id'].append(genre['id'])
    df_dict['name'].append(genre['name'])
```

```python
df_dict
```

```python
movies_genres = pd.DataFrame.from_dict(df_dict)
movies_genres
```

```python
movies_genres.to_csv('./clean_db/movies_genres.csv', index=False)
```
