```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
movies_metadata.loc[0]['production_countries']
```

```python
df_dict = {'iso_3166_1': [], 'name': []}
```

```python
columns = movies_metadata['production_countries'].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
columns
```

```python
for x, col in enumerate(columns.values):
    if not type(col) == list:
        print(col)
        columns[x] = np.NaN
```

```python
countries = []
for col in columns.values:
    if type(col) == list:
        for country in col:
            if country not in countries:
                countries.append(country)
countries
```

```python
for country in countries:
    df_dict['iso_3166_1'].append(country['iso_3166_1'])
    df_dict['name'].append(country['name'])
```

```python
df_dict
```

```python
movies_countries = pd.DataFrame.from_dict(df_dict)
movies_countries
```

```python
movies_countries.to_csv('./clean_db/movies_countries.csv', index=False)
```
