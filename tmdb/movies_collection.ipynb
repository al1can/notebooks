```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
unique_collections = movies_metadata['belongs_to_collection'].unique()
```

```python
unique_collections = pd.Series(unique_collections)
```

```python
unique_collections
```

```python
unique_collections.shape
```

```python
df_dict = {'id': [], 'name': [], 'poster_path': [], 'backdrop_path': []}
```

```python
len(movies_metadata['belongs_to_collection'])
```

```python
len(movies_metadata['belongs_to_collection'].unique())
```

```python
collections = unique_collections.apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
collections
```

```python
for x, col in enumerate(collections.values):
    if not pd.isna(col) and type((col)) != dict:
        print(col)
        collections[x] = np.NaN
```

```python
for col in collections.values:
    if not pd.isna(col) and type((col)) == dict:
        df_dict['id'].append(col['id'])
        df_dict['name'].append(col['name'])
        df_dict['poster_path'].append(col['poster_path'])
        df_dict['backdrop_path'].append(col['backdrop_path'])
```

```python
df_dict
```

```python
movies_collection = pd.DataFrame.from_dict(df_dict)
movies_collection.head()
```

```python
movies_collection.to_csv('./clean_db/movies_collection.csv', index=False)
#movies_collection.to_csv('../dataset/movies_collection.csv', index=False)
```
