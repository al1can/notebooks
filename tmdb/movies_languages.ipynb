```python
import numpy as np
import pandas as pd
import ast
```

```python
movies_metadata = pd.read_csv("../dataset/movies_metadata.csv")
pd.set_option('display.max_columns', None)
movies_metadata.head()
```

```python
df_dict = {'iso_639_1': [], 'name': []}
```

```python
columns = movies_metadata['spoken_languages'].apply(lambda x: ast.literal_eval(x) if not pd.isna(x) else x)
columns
```

```python
for col in columns.values:
        if not type(col) == list:
            print(col)
```

```python
languages = []
for col in columns.values:
    if type(col) == list:
        for language in col:
            if language not in languages:
                languages.append(language)
languages
```

```python
for language in languages:
    df_dict['iso_639_1'].append(language['iso_639_1'])
    df_dict['name'].append(language['name'])
```

```python
df_dict
```

```python
movies_languages = pd.DataFrame.from_dict(df_dict)
movies_languages
```

```python
movies_languages.to_csv('./clean_db/movies_languages.csv', index=False)
```
