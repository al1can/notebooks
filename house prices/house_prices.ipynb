```python
from sklearn.feature_selection import mutual_info_regression
```

```python
def mi_score(x, y):
    x = x.copy()
    # factorizing gives different values different numbers
    # the way ordinal encoder does i guess
    for col in x.select_dtypes(['object', 'category']):
        x[col], _ = x[col].factorize()

    # if int True else False
    disc_features = [pd.api.types.is_integer_dtype(t) for t in x.dtypes]
    mi_score = mutual_info_regression(x, y, discrete_features=disc_features, random_state=0)
    mi_score = pd.Series(mi_score, name="Mutual Info Score", index=x.columns).sort_values(ascending=False)
    
    return mi_score
```

```python
X = data.copy()
y = data.price
X.drop('price', axis=1, inplace=True)
```

```python
score = mi_score(X, y)
score
```

```python
sns.barplot(x=score.values, y=score.index)
plt.title("Mutual Scores by Columns")
```

```python
# get the top 10 column that has the highest mi score
data = data[score[:10].index.to_list()].join(y)
data
```
